```python
from rdkit import Chem
def get_pdb_sequence(pdb_id):
    try:
        pdb_folder = rxnaamapper_testset_pdb_path
        pdb_file = os.path.join(pdb_folder, f'{pdb_id}.pdb')
        mol = Chem.MolFromPDBFile(pdb_file)
        protein_sequence = Chem.MolToSequence(mol)
        return protein_sequence
    except:
        return ''
```

```python
import pymol
from tqdm import tqdm
pymol.finish_launching(['pymol', '-c'])
def handle_pdb(pdb_id):
    pymol.cmd.load(os.path.join(rxnaamapper_testset_pdb_path, '{}.pdb'.format(pdb_id)))
    pymol.cmd.remove('solvent')
    pymol.cmd.remove('resn HOH')
    pymol.cmd.remove('not polymer')
    pymol.cmd.save(os.path.join(rxnaamapper_testset_handled_pdb_path, '{}.pdb'.format(pdb_id)))
    pymol.cmd.delete('all')
```

将pdb中的多聚体分割成单体

```python
rxnaamapper_testset_handled_split_pdb_path = os.path.join(os.path.dirname(rxnaamapper_testset_handled_pdb_path), 'splited_pdb_files')
```

```python
def split_chains(pdb_id):
    pymol.cmd.load(os.path.join(rxnaamapper_testset_handled_pdb_path, '{}.pdb'.format(pdb_id)), 'multi')
    chains = pymol.cmd.get_chains()
    this_pdb_path = os.path.join(rxnaamapper_testset_handled_split_pdb_path, f'{pdb_id}')
    os.makedirs(this_pdb_path, exist_ok=True)
    for chain in chains:
        selection_name = f'{pdb_id}-{chain}'
        selection_query = f'chain {chain}'
        pymol.cmd.select(selection_name, selection_query)
        pymol.cmd.save(os.path.join(this_pdb_path, selection_name+'.pdb'), selection_name)
    pymol.cmd.delete('all')
    
```

```python
def get_split_pdb_aa_sequence(pdb_id):
    this_pdb_path = os.path.join(rxnaamapper_testset_handled_split_pdb_path, f'{pdb_id}')
    split_pdb_aa_sequence_dict = {}
    split_pdb_files = os.listdir(this_pdb_path)
    for pdb_file in split_pdb_files:
        chain_name = pdb_file.split('.')[0].split('-')[-1]
        try:
            mol = Chem.MolFromPDBFile(os.path.join(this_pdb_path ,pdb_file))
            protein_sequence = Chem.MolToSequence(mol)
        except:
            protein_sequence = ''
        split_pdb_aa_sequence_dict[chain_name] = protein_sequence
    return split_pdb_aa_sequence_dict

```

```python
rxnaamapper_testset['pdb_file_split_aa_sequence_dict'] = rxnaamapper_testset['pdb-id'].parallel_apply(lambda x:get_split_pdb_aa_sequence(x))
```

```python
def get_handled_test_set(org_aa_sequence, pdb_file_split_aa_sequence_dict, active_site):
    active_site = eval(active_site)
    mapped_results = map_active_site(org_aa_sequence, pdb_file_split_aa_sequence_dict, active_site)
    best_mapped_results = mapped_results.iloc[0].tolist()
    return best_mapped_results[:-1]
```

```python
def get_pdb_file(pdb_id, chain, is_valid):
    if is_valid:
        return f'{pdb_id}-{chain}.pdb'
    else:
        return replace_dict[pdb_id]
def get_valid_active_site_label(org_active_site, active_site, is_valid):
    if is_valid:
        return active_site
    else:
        return org_active_site
def get_valid_sequence(org_aa_sequence, aa_sequence, is_valid):
    if is_valid:
        return aa_sequence
    else:
        return org_aa_sequence
    
rxnaamapper_testset_handled_df['pdb_file'] = rxnaamapper_testset_handled_df.apply(lambda row:get_pdb_file(row['pdb-id'], row['chain'], row['is_valid_pdb']), axis=1)

rxnaamapper_testset_handled_df['vaild_active_site'] = rxnaamapper_testset_handled_df.apply(lambda row:get_valid_active_site_label(row['org_active_site'], row['active_site'], row['is_valid_pdb']), axis=1)

rxnaamapper_testset_handled_df['vaild_aa_sequence'] = rxnaamapper_testset_handled_df.apply(lambda row:get_valid_active_site_label(row['org_aa_sequence'], row['aa_sequence'], row['is_valid_pdb']), axis=1)
```
