```python
# %%
import pandas as pd
import os
import matplotlib.pyplot as plt
import seaborn as sns
# plt.rcParams['xtick.labelsize'] = 25  # 设置x轴刻度标签的字体大小
# plt.rcParams['ytick.labelsize'] = 25  # 设置y轴刻度标签的字体大小

# 定义结果路径和分数列表
include_counts = True
easifa_results_path = '../results'
baseline_results_path = './baseline_methods/baseline_results'

test_dataset_with_similarity_idx_path = '../dataset/ec_site_dataset/uniprot_ecreact_cluster_split_merge_dataset_limit_100/test_dataset_with_similarity_idx.csv'
score_list = [
    'f1_scores', 
    'mcc_scores',
    'overlap_scores', 
    'false_positive_rates', 
]
score_to_name = {
    'mcc_scores': 'MCC (higher better)',
    'f1_scores': 'F1 (higher better)',
    'false_positive_rates': 'FPR (lower better)',
    'overlap_scores': 'Recall (higher better)'
}
model_name_to_path = {
    'EasIFA-SaProt-bin (SwissProt E-RXN ASA 44,341 Enzyme)': (easifa_results_path, 'active-site-position-prediction_saprot_results.csv'),
    'EasIFA-ESM-bin (SwissProt E-RXN ASA 44,341 Enzyme)': (easifa_results_path, 'active-site-position-prediction_results.csv'),
    'BLASTp (SwissProt E-RXN ASA 44,341 Enzyme)':(baseline_results_path, 'localblastp_alignment.csv'),
    'BLASTp (All SwissProt 569,516 Protein)':(baseline_results_path, 'uniprotblastp_alignment.csv'),
    'Schrodinger-SiteMap':(baseline_results_path, 'schrodinger_sitemap.csv'),
    "AEGAN (225 samples overlapping with AEGAN's training set)":(baseline_results_path, 'AEGAN.csv'),
}

model_names = ['EasIFA-SaProt-bin (SwissProt E-RXN ASA 44,341 Enzyme)', 'EasIFA-ESM-bin (SwissProt E-RXN ASA 44,341 Enzyme)', 'BLASTp (SwissProt E-RXN ASA 44,341 Enzyme)', 'BLASTp (All SwissProt 569,516 Protein)', "AEGAN (225 samples overlapping with AEGAN's training set)", 'Schrodinger-SiteMap']


def load_similar_df(similar_df_path, include_counts=False):
    similar_df = pd.read_csv(similar_df_path)
    # Categorize 'max_tmscore'
    similar_df.loc[similar_df['max_tmscore'] > 0.5, 'max_tmscore_range'] = '0.5~1'
    similar_df.loc[(similar_df['max_tmscore'] <= 0.5) & (similar_df['max_tmscore'] > 0.2), 'max_tmscore_range'] = '0.2~0.5'
    similar_df.loc[similar_df['max_tmscore'] <= 0.2, 'max_tmscore_range'] = '0~0.2'
    


    # Add counts to the 'similarity_index_level_5' if the option is True
    if include_counts:
        # Calculate the counts for each category
        counts_sil = similar_df['similarity_index_level_5'].value_counts()
        # Define a new column that combines the 'similarity_index_level_5' with counts
        similar_df['similarity_index_level_5'] = similar_df['similarity_index_level_5'].apply(
            lambda x: f"{x} [{counts_sil[x]}]" if x in counts_sil else x
        )
        # Calculate the counts for each category
        counts_tm = similar_df['max_tmscore_range'].value_counts()
        # Define a new column that combines the 'similarity_index_level_5' with counts
        similar_df['max_tmscore_range'] = similar_df['max_tmscore_range'].apply(
            lambda x: f"{x} [{counts_tm[x]}]" if x in counts_tm else x
        )
    return similar_df



similar_df = load_similar_df(similar_df_path=test_dataset_with_similarity_idx_path, include_counts=include_counts)
if include_counts:
    max_tmscore_range_cls = list(set(similar_df['max_tmscore_range']))
    max_tmscore_range_cls.sort(key=lambda x:float(x.split('~')[0]))
    similarity_index_level_cls = list(set(similar_df['similarity_index_level_5']))
    similarity_index_level_cls.sort(key=lambda x:float(x.split('~')[0]))
    print(max_tmscore_range_cls)
    print(similarity_index_level_cls)
# 读取每个模型的结果文件，并添加一个标识列
def load_results(model_name, results_folder, results_fname, similar_df=None, add_cols=None):
    # Read the results file

    df = pd.read_csv(os.path.join(results_folder, results_fname))
    if similar_df is not None:
        add_cols_list = score_list if add_cols is None else score_list + add_cols
        df = pd.concat([similar_df, df[add_cols_list]], axis=1)

    # Add model name as a column
    df['model'] = model_name
    return df

# 加载所有模型的结果
all_results = pd.concat([
    load_results(model_name, model_name_to_path[model_name][0], model_name_to_path[model_name][1], similar_df=similar_df) 
for model_name in model_names])
all_results['model'] = pd.Categorical(all_results['model'], categories=model_names, ordered=True)

# 查看合并后的数据
# # # # display(all_results.head())

# 分组计算平均值
mean_scores_similarity_index_level = all_results.groupby(['similarity_index_level_5', 'model']).agg({score: 'mean' for score in score_list}).reset_index()


# 展示平均分数DataFrame
display(mean_scores_similarity_index_level)

# 设置画图风格
color_palette = {
    'EasIFA-SaProt-bin (SwissProt E-RXN ASA 44,341 Enzyme)': '#C76E67',  
    'EasIFA-ESM-bin (SwissProt E-RXN ASA 44,341 Enzyme)': '#1f77b4',  # 深蓝色
    'BLASTp (SwissProt E-RXN ASA 44,341 Enzyme)': '#ff7f0e',      # 橙色
    'BLASTp (All SwissProt 569,516 Protein)': '#FFD9B7',          # 淡橙色
    "AEGAN (225 samples overlapping with AEGAN's training set)": '#D5F0FF',  # 浅蓝色
    'Schrodinger-SiteMap': '#2ca02c'                             # 绿色
}
sns.set_theme(style='ticks')
sns.set_palette(color_palette.values())

# 创建一个图并绘制所有分数
fig, axes = plt.subplots(2, 2, figsize=(16, 12), sharex=True)
axes = axes.flatten()

for i, score in enumerate(score_list):
    score_name = ' '.join([x.capitalize() for x in score.split('_')]) if score not in score_to_name else score_to_name[score]
    ax = sns.barplot(data=mean_scores_similarity_index_level, x='similarity_index_level_5', y=score, hue='model', ax=axes[i], palette=color_palette, errorbar=None)
    # 设置柱状图透明度
    # for container in ax.containers:
    #     labels = [item for item in container.get_label()]
    #     if 'BLASTp (All SwissProt 569,516 Protein)' in labels:
    #         for patch in container:
    #             patch.set_alpha(0.1)  # 设置透明度为50%
    # Add annotations
    for p in ax.patches:
        height = p.get_height()
        ax.annotate(f'{height:.4f}', (p.get_x() + p.get_width() / 2., height), ha='center', va='center', fontsize=12, color='black', xytext=(0, 28), textcoords='offset points', rotation=90)
        # if p.get_facecolor() in [ax.patches[j].get_facecolor() for j in [4, 5]]:  # Indices for BLASTp (All SwissProt) and AEGAN
        #     p.set_alpha(0.1)  # Set transparency  
    axes[i].set_title('{}'.format(score_name), fontsize=20, pad=50, fontweight='bold')
    axes[i].set_xlabel('Enzyme Sequence Identity', fontsize=18)
    axes[i].set_ylabel(score_name.split(' (')[0], fontsize=18)
    axes[i].spines['top'].set_visible(False)
    axes[i].spines['right'].set_visible(False)
    axes[i].yaxis.grid(False)
    axes[i].tick_params(axis='both', labelsize=12)
    handles, labels = axes[0].get_legend_handles_labels()
    # 仅在第一个子图中显示图例
    if i == 0:
        axes[i].legend(title='Model')
        handles, labels = axes[0].get_legend_handles_labels()


    axes[i].get_legend().remove()

# 统一图例

fig.legend(handles, labels, loc='upper center', ncol=2, title='Model', frameon=False, fontsize='16', title_fontsize='16', bbox_to_anchor=(0.5, 1.10))
plt.tight_layout(rect=[0, 0, 1, 0.95])
plt.show()




```

```python
# %%
import matplotlib.pyplot as plt
from sklearn.metrics import roc_curve, precision_recall_curve
import numpy as np
from sklearn.metrics import roc_curve, precision_recall_curve, auc




# def plot_roc_pr_curves(dataframe, cut_cnt=20):
#     # 根据similarity_index_level的不同值分配颜色
#     levels = dataframe['similarity_index_level'].unique()
#     color_map = plt.cm.get_cmap('tab20', len(levels))
#     color_dict = {level: color_map(i) for i, level in enumerate(levels)}
    
#     fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(15, 5))
#     cnt = 0
#     for index, row in dataframe.iterrows():
#         prob_vector = eval(row['predict_active_prob'])  if isinstance(row['predict_active_prob'], str) else row['predict_active_prob'] # 预测的概率向量
#         true_vector = eval(row['ground_truth_active_label'])  if isinstance(row['ground_truth_active_label'], str) else row['ground_truth_active_label']# 真实的标签向量
#         prob_vector = np.array(prob_vector)
#         prob_vector = prob_vector[:, 1]
#         level = row['similarity_index_level']
#         color = color_dict[level]

#         # 计算并绘制ROC曲线
#         fpr, tpr, _ = roc_curve(true_vector, prob_vector)
#         ax1.plot(fpr, tpr, color=color, label=f'Sample {index+1} (Level {level})')

#         # 计算并绘制PR曲线
#         precision, recall, _ = precision_recall_curve(true_vector, prob_vector)
#         ax2.plot(recall, precision, color=color, label=f'Sample {index+1} (Level {level})')
#         if cnt >= cut_cnt:
#             break
#         cnt += 1

#     ax1.plot([0, 1], [0, 1], 'k--')  # 对角虚线
#     ax1.set_title('ROC Curves')
#     ax1.set_xlabel('False Positive Rate')
#     ax1.set_ylabel('True Positive Rate')
#     ax1.legend(loc='upper left', bbox_to_anchor=(1,1))

#     ax2.set_title('Precision-Recall Curves')
#     ax2.set_xlabel('Recall')
#     ax2.set_ylabel('Precision')
#     ax2.legend(loc='upper left', bbox_to_anchor=(1,1))

#     plt.tight_layout()
#     plt.show()
col_name_mapping = {
    'similarity_index_level_5': 'E.S.I.',
    'max_tmscore_range': 'Max TM-score Range',
}

def plot_average_roc_pr_curves(dataframe, col_name='similarity_index_level_5', level_sort=None):
    # col_name_mapping = {}  # Define your column name mappings if any
    show_col_name = ' '.join([x.capitalize() for x in col_name.split('_')]) if col_name not in col_name_mapping else col_name_mapping[col_name]
    levels = dataframe[col_name].unique() if level_sort is None else level_sort
    color_map = plt.cm.get_cmap('viridis', len(levels))
    color_dict = {level: color_map(i) for i, level in enumerate(levels)}

    # 准备整个样本集的平均ROC和PR曲线计算
    mean_fpr = np.linspace(0, 1, 100)
    all_tprs = []
    all_precisions = []
    all_recalls = np.linspace(0, 1, 100)
    
    # 设定子图为正方形，并保持整个图的宽高比适当
    fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(12, 6))

    for level in levels:
        level_data = dataframe[dataframe[col_name] == level]
        tprs = []
        precisions = []

        for _, row in level_data.iterrows():
            prob_vector = eval(row['predict_active_prob']) if isinstance(row['predict_active_prob'], str) else row['predict_active_prob']
            true_vector = eval(row['ground_truth_active_label']) if isinstance(row['ground_truth_active_label'], str) else row['ground_truth_active_label']
            prob_vector = np.array(prob_vector)
            prob_vector = prob_vector[:, 1] if prob_vector.ndim > 1 else prob_vector

            # 计算ROC和PR曲线数据
            fpr, tpr, _ = roc_curve(true_vector, prob_vector)
            tprs.append(np.interp(mean_fpr, fpr, tpr))
            tprs[-1][0] = 0.0  # 保证起点始终为0

            precision, recall, _ = precision_recall_curve(true_vector, prob_vector)
            precisions.append(np.interp(all_recalls, recall[::-1], precision[::-1]))

        # 平均ROC和PR曲线
        mean_tpr = np.mean(tprs, axis=0)
        mean_tpr[-1] = 1.0  # 保证终点始终为1
        roc_auc = auc(mean_fpr, mean_tpr)
        mean_precision = np.mean(precisions, axis=0)
        pr_auc = auc(all_recalls, mean_precision)

        ax1.plot(mean_fpr, mean_tpr, color=color_dict[level], linestyle='-', label=f'{show_col_name} {level} (AUC={roc_auc:.4f})', alpha=0.7, linewidth=1.5)
        ax2.plot(all_recalls, mean_precision, color=color_dict[level], linestyle='-', label=f'{show_col_name} {level} (AUC={pr_auc:.4f})', alpha=0.7, linewidth=1.5)

        all_tprs.append(mean_tpr)
        all_precisions.append(mean_precision)
    # Add F1-Score contour lines
    f1_scores = np.linspace(0.1, 0.9, 9)  # From 0.1 to 0.9
    for f1 in f1_scores:
        x_vals = np.linspace(0.0001, 1, 100)
        y_vals = f1 * x_vals / (2 * x_vals - f1)
        valid = y_vals >= 0  # Only keep valid values
        ax2.plot(x_vals[valid], y_vals[valid], color='grey', alpha=0.3, linestyle='-', linewidth=0.5)
        # Add annotation near the end of each line
        valid_x_vals = x_vals[valid]
        valid_y_vals = y_vals[valid]
        annotate_x = valid_x_vals[-11]  # 选择一个适当的点进行标注
        annotate_y = valid_y_vals[-11]
        ax2.annotate(f'F1={f1:.1f}', (annotate_x, annotate_y), textcoords="offset points", xytext=(0,10), ha='center', color='grey')
    # 设置坐标轴框为灰色并变细
    for ax in [ax1, ax2]:
        for spine in ax.spines.values():
            spine.set_color('gray')
            spine.set_linewidth(0.5)
            
    if col_name == 'similarity_index_level_5':
        # 计算并绘制所有样本的平均ROC和PR曲线
        mean_tpr = np.mean(all_tprs, axis=0)
        mean_precision = np.mean(all_precisions, axis=0)
        overall_roc_auc = auc(mean_fpr, mean_tpr)
        overall_pr_auc = auc(all_recalls, mean_precision)
        ax1.plot(mean_fpr, mean_tpr, color='black', linewidth=1.5, label=f'All Test Samples [{len(dataframe)}] Average ROC (AUC={overall_roc_auc:.4f})')
        ax2.plot(all_recalls, mean_precision, color='black', linewidth=1.5, label=f'All Test Samples [{len(dataframe)}] Average PR (AUC={overall_pr_auc:.4f})')

    ax1.plot([0, 1], [0, 1], 'k--', alpha=0.3)  # 对角虚线
    ax1.set_title('Average ROC Curve')
    ax1.set_xlabel('False Positive Rate')
    ax1.set_ylabel('True Positive Rate')
    ax1.set_xlim((0, 1))
    ax1.set_ylim((0, 1))

    ax2.set_title('Average Precision-Recall Curve')
    ax2.set_xlabel('Recall')
    ax2.set_ylabel('Precision')
    ax2.set_xlim((0, 1))
    ax2.set_ylim((0, 1))

    ax1.legend(loc='lower right',
               fontsize='10')
    ax2.legend(loc='lower left',
               fontsize='10')

    plt.tight_layout()
    plt.show()


def plot_average_roc_pr_curves_for_ablation_study(dataframe, col_name='model', level_sort=None, model_name_mapping=None):
    # col_name_mapping = {}  # Define your column name mappings if any
    
    model_names = dataframe[col_name].unique() if level_sort is None else level_sort
    color_map = plt.cm.get_cmap('viridis', len(model_names))
    color_dict = {level: color_map(i) for i, level in enumerate(model_names)}

    # 准备整个样本集的平均ROC和PR曲线计算
    mean_fpr = np.linspace(0, 1, 100)
    all_tprs = []
    all_precisions = []
    all_recalls = np.linspace(0, 1, 100)
    
    # 设定子图为正方形，并保持整个图的宽高比适当
    fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(12, 6))

    for model in model_names:
        level_data = dataframe[dataframe[col_name] == model]
        tprs = []
        precisions = []

        for _, row in level_data.iterrows():
            prob_vector = eval(row['predict_active_prob']) if isinstance(row['predict_active_prob'], str) else row['predict_active_prob']
            true_vector = eval(row['ground_truth_active_label']) if isinstance(row['ground_truth_active_label'], str) else row['ground_truth_active_label']
            prob_vector = np.array(prob_vector)
            prob_vector = prob_vector[:, 1] if prob_vector.ndim > 1 else prob_vector

            # 计算ROC和PR曲线数据
            fpr, tpr, _ = roc_curve(true_vector, prob_vector)
            tprs.append(np.interp(mean_fpr, fpr, tpr))
            tprs[-1][0] = 0.0  # 保证起点始终为0

            precision, recall, _ = precision_recall_curve(true_vector, prob_vector)
            precisions.append(np.interp(all_recalls, recall[::-1], precision[::-1]))

        # 平均ROC和PR曲线
        mean_tpr = np.mean(tprs, axis=0)
        mean_tpr[-1] = 1.0  # 保证终点始终为1
        roc_auc = auc(mean_fpr, mean_tpr)
        mean_precision = np.mean(precisions, axis=0)
        pr_auc = auc(all_recalls, mean_precision)
        show_model_name = model_name_mapping[model] if model_name_mapping is not None else model
        ax1.plot(mean_fpr, mean_tpr, color=color_dict[model], linestyle='-', label=f'{show_model_name} (AUC={roc_auc:.4f})', linewidth=1.5)
        ax2.plot(all_recalls, mean_precision, color=color_dict[model], linestyle='-', label=f'{show_model_name} (AUC={pr_auc:.4f})', linewidth=1.5)

        all_tprs.append(mean_tpr)
        all_precisions.append(mean_precision)
    # Add F1-Score contour lines
    f1_scores = np.linspace(0.1, 0.9, 9)  # From 0.1 to 0.9
    for f1 in f1_scores:
        x_vals = np.linspace(0.0001, 1, 100)
        y_vals = f1 * x_vals / (2 * x_vals - f1)
        valid = y_vals >= 0  # Only keep valid values
        ax2.plot(x_vals[valid], y_vals[valid], color='grey', alpha=0.3, linestyle='-', linewidth=0.5)
        # Add annotation near the end of each line
        valid_x_vals = x_vals[valid]
        valid_y_vals = y_vals[valid]
        annotate_x = valid_x_vals[-11]  # 选择一个适当的点进行标注
        annotate_y = valid_y_vals[-11]
        ax2.annotate(f'F1={f1:.1f}', (annotate_x, annotate_y), textcoords="offset points", xytext=(0,10), ha='center', color='grey')
    # 设置坐标轴框为灰色并变细
    for ax in [ax1, ax2]:
        for spine in ax.spines.values():
            spine.set_color('gray')
            spine.set_linewidth(0.5)
            
    ax1.plot([0, 1], [0, 1], 'k--', alpha=0.3)  # 对角虚线
    ax1.set_title('Average ROC Curve')
    ax1.set_xlabel('False Positive Rate')
    ax1.set_ylabel('True Positive Rate')
    ax1.set_xlim((0, 1))
    ax1.set_ylim((0, 1))

    ax2.set_title('Average Precision-Recall Curve')
    ax2.set_xlabel('Recall')
    ax2.set_ylabel('Precision')
    ax2.set_xlim((0, 1))
    ax2.set_ylim((0, 1))

    ax1.legend(loc='lower right',
               fontsize='10'
               )
    ax2.legend(loc='lower left',
               fontsize='10'
               )
    

    plt.tight_layout()
    plt.show()


def plot_average_roc_pr_curves_for_ablation_reaction_branch(dataframe, model_col='model', similarity_col='similarity_index_level_5', level_sort=None, model_name_mapping=None):
    bar_color_dict = {
    'EasIFA-ESM-bin': '#1f77b4',  # 深蓝色
    'EasIFA-E-bin': '#8c564b',
    }
    # 模型名称和相似度级别
    model_names = dataframe[model_col].unique()
    levels = dataframe[similarity_col].unique() if level_sort is None else level_sort
    level_col_name = col_name_mapping[similarity_col]
    
    # 为每种模型和每种相似度等级生成不同的颜色和线型
    color_map = plt.cm.get_cmap('viridis', len(levels))
    line_styles = ['-', '--']
    color_dict = {level: color_map(i) for i, level in enumerate(levels)}
    
    # 准备绘图
    fig = plt.figure(figsize=(12, 12))
    ax1 = fig.add_subplot(2, 2, 1)  # 第一行，第一列
    ax2 = fig.add_subplot(2, 2, 2)  # 第一行，第二列
    ax3 = fig.add_subplot(2, 1, 2)  # 第二行，跨越两列
    # ax1.set_aspect('equal', adjustable='box')
    # ax2.set_aspect('equal', adjustable='box')

    pr_auc_list = []



    # 逐一处理每种模型和相似度等级
    for model in model_names:
        for level in levels:
            # 筛选特定模型和相似度级别的数据
            level_data = dataframe[(dataframe[model_col] == model) & (dataframe[similarity_col] == level)]
            tprs = []
            precisions = []
            mean_fpr = np.linspace(0, 1, 100)
            all_recalls = np.linspace(0, 1, 100)

            for _, row in level_data.iterrows():
                prob_vector = eval(row['predict_active_prob']) if isinstance(row['predict_active_prob'], str) else row['predict_active_prob']
                true_vector = eval(row['ground_truth_active_label']) if isinstance(row['ground_truth_active_label'], str) else row['ground_truth_active_label']
                prob_vector = np.array(prob_vector)
                prob_vector = prob_vector[:, 1] if prob_vector.ndim > 1 else prob_vector

                # 计算ROC和PR曲线数据
                fpr, tpr, _ = roc_curve(true_vector, prob_vector)
                tprs.append(np.interp(mean_fpr, fpr, tpr))
                tprs[-1][0] = 0.0

                precision, recall, _ = precision_recall_curve(true_vector, prob_vector)
                precisions.append(np.interp(all_recalls, recall[::-1], precision[::-1]))

            # 计算平均ROC和PR曲线
            mean_tpr = np.mean(tprs, axis=0)
            mean_tpr[-1] = 1.0
            roc_auc = auc(mean_fpr, mean_tpr)
            mean_precision = np.mean(precisions, axis=0)
            pr_auc = auc(all_recalls, mean_precision)
            # pr_auc_dict[model].append(pr_auc)
            pr_auc_list.append((model, level, pr_auc))

            # 获取模型名称和级别的显示名称
            show_model_name = model_name_mapping[model] if model_name_mapping else model
 
            linestyle = line_styles[model_names.tolist().index(model) % len(line_styles)]
            
            ax1.plot(mean_fpr, mean_tpr, label=f'{show_model_name} - {level_col_name} {level} (AUC={roc_auc:.4f})', color=color_dict[level], linestyle=linestyle)
            ax2.plot(all_recalls, mean_precision, label=f'{show_model_name} - {level_col_name} {level} (AUC={pr_auc:.4f})', color=color_dict[level], linestyle=linestyle)
            


    pr_auc_df = pd.DataFrame(pr_auc_list)
    pr_auc_df.columns = ['model', 'similarity_index_level_5', 'prauc']
    
    # 添加柱状图
    # for i, model in enumerate(model_names):
    #     pr_aucs = pr_auc_dict[model]
    #     ax3.bar(np.arange(len(pr_aucs)) + i * 0.2, pr_aucs, width=0.2, color=color_dict[model], label=f'model {model}')
    sns.barplot(data=pr_auc_df, x='similarity_index_level_5', y='prauc', hue='model', ax=ax3, palette=bar_color_dict, errorbar=None, width=0.5)
    for level in levels:
        subset = pr_auc_df[pr_auc_df['similarity_index_level_5'] == level]
        if len(subset['model'].unique()) > 1:
            y_values = subset['prauc'].tolist()
            ax3.text(subset.index[0] + 0.2, max(y_values) - 0.01, f'Δ={abs(y_values[1] - y_values[0]):.4f}', ha='center', va='bottom', color='black')
    ax3.set_title('PR AUC Difference')
    ax3.set_xlabel('Enzyme Sequence Identity')
    ax3.set_ylabel('PR AUC')
    ax3.set_ylim((0.65, 1))

    # ax3.set_aspect('equal', adjustable='box') 


    # Add F1-Score contour lines
    f1_scores = np.linspace(0.1, 0.9, 9)  # From 0.1 to 0.9
    for f1 in f1_scores:
        x_vals = np.linspace(0.0001, 1, 100)
        y_vals = f1 * x_vals / (2 * x_vals - f1)
        valid = y_vals >= 0  # Only keep valid values
        ax2.plot(x_vals[valid], y_vals[valid], color='grey', alpha=0.3, linestyle='-', linewidth=0.5)
        # Add annotation near the end of each line
        valid_x_vals = x_vals[valid]
        valid_y_vals = y_vals[valid]
        annotate_x = valid_x_vals[-11]  # 选择一个适当的点进行标注
        annotate_y = valid_y_vals[-11]
        ax2.annotate(f'F1={f1:.1f}', (annotate_x, annotate_y), textcoords="offset points", xytext=(0,10), ha='center', color='grey')
    # 设置坐标轴框为灰色并变细
    for ax in [ax1, ax2, ax3]:
        for spine in ax.spines.values():
            spine.set_color('gray')
            spine.set_linewidth(0.5)
            
    ax1.plot([0, 1], [0, 1], 'k--', alpha=0.3)  # 对角虚线
    ax1.set_title('Average ROC Curve')
    ax1.set_xlabel('False Positive Rate')
    ax1.set_ylabel('True Positive Rate')
    ax1.set_xlim((0, 1))
    ax1.set_ylim((0, 1))

    ax2.set_title('Average Precision-Recall Curve')
    ax2.set_xlabel('Recall')
    ax2.set_ylabel('Precision')
    ax2.set_xlim((0, 1))
    ax2.set_ylim((0, 1))

    # 设置图例字体大小
    ax1.legend(
        loc='lower right',
        fontsize='9'
        )
    ax2.legend(
        loc='lower left',
        fontsize='9'
        )
    ax3.legend(
        loc='upper right',
        fontsize='10'
        )

    plt.tight_layout()
    plt.show()

```
